```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url = "https://taeillo.com/products"
```

```python
r = requests.get(url).text
```

```python
soup = BeautifulSoup(r, 'lxml')
#price = soup.find_all('span', class_='money')
#product = soup.find_all('h3', class_='popup_cart_title')
product_card = soup.find_all('div', class_='single-product__content ')


with open('taeillo.csv', 'w+') as file:
  file.write('Product,Price (₦) \n')
  for i in product_card:
    product = i.div.h3.a.text
    price = i.find('span', class_='money').text.replace(',','').replace('₦', '')

    #print(i)
    file.write(product)
    file.write(',')
    file.write(price)
    file.write('\n')
```

#Converting a Json File into a CSV File

```python
url = "/content/drive/MyDrive/Python Books/products.json"
```

```python
taeillo_csv = ''''''
```

```python
with open(url, 'r+') as file:
  json_file = json.load(file)
  for i,v in enumerate(json_file['products']):
    for j in v.keys(): # First add the .keys() to the empty string then the .items()
      taeillo_csv = taeillo_csv+ j + "||"
    taeillo_csv = taeillo_csv + "\n"
    break
```

```python
with open(url, 'r+') as file:
  json_file = json.load(file)
  for i,v in enumerate(json_file['products']):
    for j in v.items(): # First add the .keys() to the empty string then the .items()
      taeillo_csv = taeillo_csv + str(j[1:]) + "||"
    taeillo_csv = taeillo_csv + "\n"
    
```

```python
#print(taeillo_csv)

taeillo_csv = taeillo_csv.replace(",)", ")").replace(",", "__").replace("||", ",").replace("(", "").replace(")", "").replace("'", "")

print(taeillo_csv)
```

```python
with open("/content/drive/MyDrive/Python Books/taeillo_csv.csv", "w+") as file:
  file.write(taeillo_csv)
```
